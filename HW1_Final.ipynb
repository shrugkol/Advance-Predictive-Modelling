# Question 2: Guess the coin (5+5 = 10 points)

Assume we are playing a guessing game with a friend. The friend has three coins at hand:
* **Coin A**: a fair coin with equal probability of being head (H ) or tail (T)
* **Coin B**: a biased coin with a probability of 0.75 being head (H)
* **Coin C**: a coin with a probability of $P^*(H)$ being head (H)

The friend secretly picked one of the coins, randomly flipped it a few times, and get a sequence of *HTHHTT* (all of which come from the same coin he picked). 

1. If you had to choose between whether **Coin A** or **Coin B** was used, and you prefer the one which is more likely, which one will you choose?  (5 points)
2. What would be the value of  $P^*(H)$ such that **Coin C** corresponds to the most likely (among infinite possible coins) coin to have produced the observed sequence?(i.e. provide  an analytical derivation for $P^*(H)$ using maximum likelihood estimation (MLE))."

## Answer

1. The maximum likelihood principle states that we choose the estimator for the parameter $\theta$ that maximizes the likelihood of observing the actual sample given the probability density function for y. We take our observation (*HTHHTT*) and then "look back" to calculate the probability of Coin A versus Coin B resulting in this sample observation. Thus, for Coin A, our parameter is $\theta$ = 0.5. For Coin B, our parameter is $\theta$ = 0.75. Whichever value of P results in a higher probability of the sample occurring is the coin we determine was more likely used. 
    To solve this, we plug each value of P into the formula $L = (\theta)^{3}*((1-\theta)^{3})$

    $$\operatorname{Likelihood} (\operatorname{coin} A)=\operatorname{P}(H T H H T T \mid \operatorname{coin} A)=(0.5)^{6} \approx 0.0156$$
    $$\operatorname{Likelihood}(\operatorname{coin} B)=\operatorname{P}(H T H H T T \mid \operatorname{coin} B)=(0.75)^{3} \times(0.25)^{3} \approx 0.00659$$
    $$\operatorname{Likelihood} (\operatorname{coin} A)>\operatorname{Likelihood}(\operatorname{coin} B)$$

    **Coin A has a higher likelihood.** This makes sense because Heads and Tails are equally present in the sample, so the fair coin is more likely to have produced this sample.


2. The value of $P^*(H)$ that corresponds to the most likely coin to have produced the observed sequence is the value that maximizes the function $L = (\theta)^{3}*(1-\theta)^{3}$ subject to the constraint that P is between 0 and 1.In order to find the maximum value of this function, we take the , the natural logarithm of the likelihood function
    
$(logL(\theta) = 3\log\theta + 3\log(1-\theta)$

    
   We then solve for the maximum value of the function by setting $\frac{dL}{d\theta}$ equal to 0. 
    
    
$\frac{dL}{d\theta} = 3/\theta - 3/(1-\theta)$

$0 = 3/\theta- 3/(1-\theta)$ 

$3/\theta = 3/(1-\theta)$

$1/\theta = 1/(1-\theta)$

$\theta = 1-\theta $
  
$\theta = \frac{1}{2}$

Thus, the value of $P^*(H)$ such that Coin C corresponds to the most likely coin to have produced this sequence is **0.5.**

```python
import numpy as np
from sklearn.model_selection import KFold, GridSearchCV

alphas =  10**np.linspace(10,-2,100)*0.5
kr = KFold(n_splits=5)

lasso = Lasso(random_state=0, max_iter=10000)
ridge = Ridge(random_state=0, max_iter=10000)

tuned_parameters = [{'alpha': alphas}]

clf = GridSearchCV(lasso, tuned_parameters, cv=kr, refit=False, scoring = 'neg_mean_absolute_error')
clf.fit(X_train_scaled, y_train)
print('Lasso')
print(clf.best_params_)
alpha_L = clf.best_params_
#for lasso, 0.0815

crf = GridSearchCV(ridge, tuned_parameters, cv=kr, refit=False, scoring = 'neg_mean_absolute_error')
crf.fit(X_train_scaled, y_train)
print('\nRidge')
print(crf.best_params_)
alpha_R = crf.best_params_

#for ridge, 0.0267
```

2.

```python
lasso = Lasso(max_iter = 10000, normalize = True)
coefs_L = []

for a in alphas:
    lasso.set_params(alpha=a)
    lasso.fit(X_train_scaled, y_train)
    coefs_L.append(lasso.coef_)

plt.subplot(1,2,1)
ax = plt.gca()
ax.plot(alphas*2, coefs_L)
ax.set_xscale('log')
plt.axis('tight')
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('Lasso')



ridge = Ridge(max_iter = 10000, normalize = True)
coefs_R = []

for a in alphas:
    ridge.set_params(alpha=a)
    ridge.fit(X_train_scaled, y_train)
    coefs_R.append(ridge.coef_)

    
plt.subplot(1,2,2)
ax2 = plt.gca()
ax2.plot(alphas*2, coefs_R)
ax2.set_xscale('log')
plt.axis('tight')
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('Ridge')

plt.show()
```

As the value of the regularization parameter increases, the number of variables in the model for Lasso approaches 0 as the Lasso model can set variables exactly equal to 0. As the value of the regularization parameter increases, the Ridge coefficients shrink towards 0.
